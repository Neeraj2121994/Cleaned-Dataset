```python
def handlerate(value):
    if value=='NEW' or value=='-':
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
data_set['rate']=data_set['rate'].apply(handlerate)
data_set.head()
    
```

```python
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return(float(value))
            
```

```python
data_set['cost2plates']=data_set['cost2plates'].apply(handlecomma)
data_set.head()
```

```python
rest_types=data_set['rest_type'].value_counts(ascending=False)
rest_types
```

```python
rest_types_lt_1000=rest_types[rest_types<1000]
rest_types_lt_1000
```

```python
def handle_rest_type(value):
    if(value in rest_types_lt_1000):
        return 'others'
    else:
        return value
data_set['rest_type']=data_set['rest_type'].apply(handle_rest_type)
data_set['rest_type'].unique()
data_set['rest_type'].value_counts()
```

```python
location=data_set['location'].value_counts()
location
```

```python
location_lh_300=location[location<300]
location_lh_300
```

```python
def location_combine(value):
    if value in (location_lh_300):
        return 'others'
    else:
        return value
data_set['location']=data_set['location'].apply(location_combine)
data_set['location'].value_counts()
```

```python
cuisines=data_set['cuisines'].value_counts()
```

```python
cuisines_lt_100=cuisines[cuisines<100]
cuisines_lt_100
```

```python
def cuisines_combine(value):
    if value in cuisines_lt_100:
        return 'others'
    else:
        return value
data_set['cuisines']=data_set['cuisines'].apply(cuisines_combine)
data_set['cuisines'].value_counts()
```
